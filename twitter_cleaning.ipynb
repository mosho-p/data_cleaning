#### I have 6 json files, I need to see if the keys are different:

```python
def preview_data(file_name):
    """Print metadata about the file, then return the list of columns for comparison with other files"""
    with open('data/'+file_name) as f:
        print(file_name)
        print('Number of tweets:', str(sum([1 for x in f.readlines()])))
        f.seek(0)
        return set(json.loads(f.readline()).keys())
```

```python
json_files = sorted([x for x in os.listdir('data/') if re.match(r'.\.no-labels\.json', x)])
print(*json_files, sep='\n')   # all of the twitter json files
```

```python
keys = set(df_a.columns)
for file in json_files:
    x = preview_data(file)
    if x-keys:  # column in one file and not in others
        print(x-keys)
    keys.update(x)
    print()
```

#### I can work with all of this data in memory. There are some files with extra keys.

```python
df_book = dict()
for file in json_files:
    df_book[file[0]+'_df'] = pd.read_json('data/'+file, lines=True)
```

```python
full_df = pd.concat(df_book.values(), axis=0, ignore_index=True, sort=False)
```

```python
summary(full_df)
```

#### Now I can drop contributors with confidence

```python
def func_on_all_dfs(func, keys=['a_df', 'b_df', 'e_df', 'f_df', 'g_df', 'h_df']):
    """
    Perform transformation functions in place on all of the separate dataframes.
    param keys (list): specify dataframes to change. For example, if some dataframes are missing a column.
    """
    global full_df, df_book
    full_df = func(full_df)
    for k in keys:
        df_book[k] = func(df_book[k])
```

```python
func_on_all_dfs(lambda x: x.drop(columns='contributors'))
```
